```python
def load_transform_data(url, agg_func="mean"):

    datapi = pd.read_table(url, delim_whitespace=True, header=None)

    for i in range(len(datapi[0])):
        if len(str(datapi[0][i])) == 8:
            ds = "00" + str(datapi[0][i])
            datapi.at[i, 0] = ds
        elif len(str(datapi[0][i])) == 9:
            ds = "0" + str(datapi[0][i])
            datapi.at[i, 0] = ds
        elif len(str(datapi[0][i])) == 10:
            datapi.drop(i, inplace=True)

    state = datapi.iloc[:, 0].astype(str).str[:3]
    year = datapi.iloc[:, 0].astype(str).str[-4:]

    if agg_func == "mean":
        variable = datapi.iloc[:, 1:].mean(axis=1)
    elif agg_func == "min":
        variable = datapi.iloc[:, 1:].min(axis=1)
    elif agg_func == "max":
        variable = datapi.iloc[:, 1:].max(axis=1)
    else:
        raise ValueError("Invalid aggregation function. Use 'mean', 'min', or 'max'.")

    my_data = pd.DataFrame({
        "state": state,
        "year": year,
        agg_func: variable
    })

    my_data[agg_func] = my_data[agg_func].astype(int).apply(lambda x: round((x - 32) * 5/9))

    return my_data
```
